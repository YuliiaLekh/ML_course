1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = sp.symbols('x')

f = 10 * x - 3
derivative_f = sp.diff(f, x)

f_func = sp.lambdify(x, f, modules='numpy')
derivative_f_func = sp.lambdify(x, derivative_f, modules='numpy')

x_vals = np.linspace(0, 10, 400)
f_vals = f_func(x_vals)
derivative_f_vals = np.full_like(x_vals, derivative_f_func(0))

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = 10x - 5')
plt.plot(x_vals, derivative_f_vals, '--', label="f'(x) = 10")
plt.title('Графіки функції та її похідної')
plt.xlabel('Кількість годин навчання (x)')
plt.ylabel('Оцінка')
plt.legend()
plt.grid(True)
plt.show()
```

На графіку видно, що кількість годин витрачених на навчання впливає на оцінку. Чим більше годин, тим вища оцінка

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.ln(x)

f1_prime = sp.diff(f1, x) 
f2_prime = sp.diff(f2, x)

f1_func = sp.lambdify(x, f1, modules='numpy')
f1_prime_func = sp.lambdify(x, f1_prime, modules='numpy')
f2_func = sp.lambdify(x, f2, modules='numpy')
f2_prime_func = sp.lambdify(x, f2_prime, modules='numpy')

x_vals_f1 = np.linspace(-2*np.pi, 2*np.pi, 300)
x_vals_f2 = np.linspace(0.1, 10, 300)

f1_vals = f1_func(x_vals_f1)
f1_prime_vals = f1_prime_func(x_vals_f1)

f2_vals = f2_func(x_vals_f2)
f2_prime_vals = f2_prime_func(x_vals_f2)

# sin(x) та похідна
plt.figure(figsize=(10, 4))
plt.plot(x_vals_f1, f1_vals, label='f(x) = sin(x)')
plt.plot(x_vals_f1, f1_prime_vals, '--', label="f'(x) = cos(x)")
plt.title('Функція sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()

# ln(x) та похідна
plt.figure(figsize=(10, 4))
plt.plot(x_vals_f2, f2_vals, label='f(x) = ln(x)')
plt.plot(x_vals_f2, f2_prime_vals, '--', label="f'(x) = 1/x")
plt.title('Функція ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1
f_prime = sp.diff(f, x)

f_func = sp.lambdify(x, f, modules='numpy')
f_prime_func = sp.lambdify(x, f_prime, modules='numpy')

x_vals = np.linspace(0, 100, 400)

f_vals = f_func(x_vals)
f_prime_vals = f_prime_func(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = x² - 5x + 1')
plt.plot(x_vals, f_prime_vals, '--', label="f'(x) = 2x - 5")
plt.title('Функція загальних витрат і її похідна')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Витрати')
plt.legend()
plt.grid(True)
plt.show()
```
